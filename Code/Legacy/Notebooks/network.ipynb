```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import osmnx as ox
```

```python
moda=ox.graph_from_place('Moda,Kadik√∂y,Istanbul', network_type='drive')
```

```python
moda_nodes=list(moda.nodes())
```

```python
moda_nodes
```

```python
moda_node_name_dict={}

for i in range(len(moda_nodes)):
    moda_node_name_dict[i]=moda_nodes[i]
```

```python
ox.plot_graph(moda)
```

```python
adjacency_matrix = nx.to_numpy_array(moda,weight='length')
```

```python
adjacency_matrix
```

```python
G=nx.DiGraph()
```

```python

for i in range(len(adjacency_matrix)):
    for j in range(len(adjacency_matrix)):
        if adjacency_matrix[i][j]!=0:
            G.add_edge(i,j,weight=adjacency_matrix[i][j])
            
```

```python
path=nx.shortest_path(G,source=0,target=5,weight='weight')

```

```python
path_moda=[]
for i in path:
    path_moda.append(moda_node_name_dict[i])
    
```

```python
ox.plot_graph_route(moda,path_moda,)
```
